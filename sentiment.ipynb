```python
texts = [
    "I love this product",
    "This is the worst experience",
    "Amazing quality and service",
    "I hate this item",
    "Very happy with the purchase",
    "Not worth the money"
]

labels = [1, 0, 1, 0, 1, 0]
#TF-IDF
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer(lowercase=True)
x_text = vector.fit_transform(texts)

model_nb = MultinomialNB()
model_nb.fit(x_text,labels)
test_txt = ["i love this product"]
test_x = vector.transform(test_txt)
predict = model_nb.predict(test_x)

print("positive" if predict[0]==1 else "negative")
```
